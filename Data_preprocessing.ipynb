```python
import pandas as pd

df = pd.read_csv('Patient-Medical-Data-for-Novel-Coronavirus-COVID-19.csv')
```

```python
import numpy as np
df = df.replace('Missing["NotAvailable"]', np.nan)
df.head()
```

```python
cancer = set(['myeloma', 'colon cancer', 'prostate cancer', 'lung cancer'])
diabetes = set(['diabetes', 'diabetes mellitus', 'type 2 diabetes', 'tongue cancer'])
hypertension = set(['hypothyroidism', 'cerebrovascular infarct', 'cerebral infarction', 'cerebrovascular accident infarct', 
                'benign prostatic hypertrophy', 'hypertension', 'systemic arterial hypertension',
                'hemorrhage of digestive tract', 'obesity', 'dyslipidemia', 'stenocardia', 'atherosclerosis'])
coronary_heart_disease = set(['cardiac disease', 'coronary stent', 'cardiac dysrhythmia',
                          'frequent ventricular premature beat', 'cardiovascular disease', 'valvular heart disease',
                         'coronary heart disease', 'atrial fibrillation', 'cardiomyopathy'])
renal_disease = set(['kidney disease', 'on dialysis', 'chronic renal insufficiency', 'prerenal azotemia', 'chronic kidney disease',
                 'myocarditis'])
immunodeficiency = set(['HIV positive'])
hepatit = set(['hepatitis B'])
lang = set(['bronchial asthma', 'tuberculosis', 'asthma', 'chronic obstructive pulmonary disease', 
        'chronic bronchitis', 'lung cancer', 'chronic pulmonary condition'])
```

```python
diseases = df.dropna(subset=['ChronicDiseases']).reset_index(drop=True)
diseases_name = ['cancer', 'diabetes', 'hypertension','coronary_heart_disease','renal_disease',
                 'immunodeficiency','hepatit']
chronic = {'cancer':cancer, 'diabetes':diabetes, 'hypertension':hypertension, 
           'coronary_heart_disease':coronary_heart_disease, 'renal_disease':renal_disease,
          'immunodeficiency':immunodeficiency, 'hepatit':hepatit, 'lang':lang}
patient_data = {name: [] for name in diseases_name}

symptoms_name = ['fever', 'cough','shortness of breath','diarrhea', 'fatigue', 'vomit', 'interstitial abnormalities',
                 'ground-glass opacity']
fever = set(['fever', 'pneumonia', 'sepsis'])
cough = set(['cough', 'respiratory symptoms', 'sputum', 'sore throat'])
short = set(['acute respiratory distress syndrome', 'respiratory symptoms', 'acute respiratory disease syndrome', 
             'acute respiratory disease', 'gasp', 'shortness of breath'])
diarrhea = set(['gastritis', 'diarrhea'])
fatigue = set(['headache', 'dizziness', 'multiple electrolyte imbalance', 'fatigue', 'somnolence', 'weakness', 'chills'])
vomit = set(['nausea', 'multiple electrolyte imbalance', 'gastritis', 'vomiting'])
abnormal = set(['respiratory symptoms', 'chest pain', 'gasp', 'acute respiratory failure', 'hypoxia',
                'severe acute respiratory distress syndrome'])
ground_glass = set(['sputum', 'pneumonia'])
heart = set(['congestive heart failure', 'cardiopulmonary arrest', 'cardiac arrhythmia', 'hypertension',
             'myocardial infarction', 'arrhythmia', 'acute coronary syndrome', 'myalgias', 'cardiogenic shock',
             'heart failure', 'acute left heart failure'])
symptoms = {'fever':fever, 'cough':cough, 'shortness of breath':short, 'diarrhea':diarrhea, 'fatigue':fatigue,
            'vomit':vomit, 'interstitial abnormalities':abnormal, 'ground-glass opacity':ground_glass, 
            'heart':heart}
for name in symptoms_name:
    patient_data[name] = []
patient_data['gender'] = []
patient_data['old_age'] = []
print(diseases.shape)
```

```python
#add people with chronic diseases
diseases = df
for i in range(diseases.shape[0]):
    if pd.isna(diseases.loc[i]['ChronicDiseases']):
        continue
    if pd.isna(diseases.loc[i]['Symptoms']):
        continue
    diseases_ = diseases.loc[i]['ChronicDiseases'].replace('{', '')
    diseases_ = diseases_.replace('}', '')
    diseases_ = diseases_.replace('"', '')
    diseases_ = set(diseases_.split(', '))
    for name in diseases_name:
        patient_data[name].append(1 if len(chronic[name].intersection(diseases_)) else 0)
    symptoms_ = diseases.loc[i]['Symptoms'].replace('{', '')
    symptoms_ = symptoms_.replace('}', '')
    symptoms_ = symptoms_.replace('"', '')
    symptoms_ = set(symptoms_.split(', '))
    for name in symptoms_name:
        patient_data[name].append(1 if len(symptoms[name].intersection(symptoms_)) else 0)
    patient_data['gender'].append(1 if diseases.loc[i]['Sex'].find('Male') != -1 else 0)
    patient_data['old_age'].append(1 if diseases.loc[i]['Age'].isnumeric() and int(diseases.loc[i]['Age']) > 60 else 0)

data = pd.DataFrame(patient_data)
print(data.shape)
```

```python
#add people without chronic diseases
c = 0
for i in range(diseases.shape[0]):
    if not pd.isna(diseases.loc[i]['ChronicDiseases']):
        continue
    if pd.isna(diseases.loc[i]['Age']):
        continue
    if pd.isna(diseases.loc[i]['Symptoms']):
        continue
    c+= 1
    diseases_ = set()
    for name in diseases_name:
        patient_data[name].append(1 if len(chronic[name].intersection(diseases_)) else 0)
    symptoms_ = diseases.loc[i]['Symptoms'].replace('{', '')
    symptoms_ = symptoms_.replace('}', '')
    symptoms_ = symptoms_.replace('"', '')
    symptoms_ = set(symptoms_.split(', '))
    for name in symptoms_name:
        patient_data[name].append(1 if len(symptoms[name].intersection(symptoms_)) else 0)
    patient_data['gender'].append(1 if diseases.loc[i]['Sex'].find('Male') != -1 else 0)
    patient_data['old_age'].append(1 if diseases.loc[i]['Age'].isnumeric() and int(diseases.loc[i]['Age']) > 60 else 0)
    if c == 300:
        break
data = pd.DataFrame(patient_data)
print(data.shape)
```

```python
for col in data.columns:
    print(col, data[col].sum()/data.shape[0])
```
